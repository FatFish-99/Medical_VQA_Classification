```python
from bert_score import score
import pandas as pd
```

```python
def compute_bertscore(predictions, references, model_type='distilbert-base-uncased'):
    # Clean — convert all to string and filter out empty pairs
    cleaned = [
        (str(p), str(r)) 
        for p, r in zip(predictions, references)
        if p and r and str(p).strip() != '' and str(r).strip() != '' 
        and str(p) != 'nan' and str(r) != 'nan'
    ]
    
    if not cleaned:
        print('No valid predictions found!')
        return {'bertscore_precision': 0.0, 'bertscore_recall': 0.0, 'bertscore_f1': 0.0}
    
    preds, refs = zip(*cleaned)
    preds = list(preds)
    refs  = list(refs)
    
    print(f'Running BERTScore on {len(preds)} valid samples...')
    
    P, R, F1 = score(
        preds,
        refs,
        model_type=model_type,
        lang='en',
        verbose=True
    )
    return {
        'bertscore_precision': round(P.mean().item(), 4),
        'bertscore_recall':    round(R.mean().item(), 4),
        'bertscore_f1':        round(F1.mean().item(), 4),
    }


def print_bertscore(metrics, model_name='Model'):
    print(f"\n{'='*40}")
    print(f'  {model_name}')
    print(f"{'='*40}")
    print(f"  BERTScore P:  {metrics['bertscore_precision']:.4f}")
    print(f"  BERTScore R:  {metrics['bertscore_recall']:.4f}")
    print(f"  BERTScore F1: {metrics['bertscore_f1']:.4f}")
    print(f"{'='*40}\n")
```

```python
# BLIP-2
df_blip2 = pd.read_csv('blip/blip-2/blip2_all_predictions.csv')
metrics_blip2_bert = compute_bertscore(
    df_blip2['blip2_zeroshot'].tolist(),
    df_blip2['reference'].tolist()
)
print_bertscore(metrics_blip2_bert, 'BLIP-2 Zero-Shot')

# InstructBLIP
df_instruct = pd.read_csv('blip/instructblip/results/instructblip_all_predictions.csv')
metrics_instruct_bert = compute_bertscore(
    df_instruct['instructblip'].tolist(),
    df_instruct['reference'].tolist()
)
print_bertscore(metrics_instruct_bert, 'InstructBLIP vicuna-7b')

# GPT-5 Mini
df_gpt = pd.read_csv('gpt/gpt5mini_all_predictions.csv')
metrics_gpt_bert = compute_bertscore(
    df_gpt['gpt5mini'].tolist(),
    df_gpt['reference'].tolist()
)
print_bertscore(metrics_gpt_bert, 'GPT-5 Mini')
```
